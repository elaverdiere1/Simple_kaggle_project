Simple work on mnist data set to show good coding and comments along with knowledge of neural networks

```python
# list imports first
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras import utils
from sklearn.model_selection import train_test_split
```

```python
# pulling in and checking data
train = pd.read_csv('./train.csv')
train.head()
```

```python
#examining the data
train.shape
```

```python
# spliting the data into X and y for the train and test split
X = train.drop('label', axis=1).values
y = train['label']
```

```python
#sample of single image
X[0]
```

```python
#scaling X data to be values between 0 and 1, then making the y categorical so that it is values of 0 or 1
X = X/255
y = utils.to_categorical(y)
```

```python
#checking the x and y values to make sure they are correctly scaled
X[0]
```

```python
y[0:5]
```

```python
#setting up the train test split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42)
```

```python
# reshaping the images for the neural network
# code help https://www.datacamp.com/community/tutorials/cnn-tensorflow-python
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)
```

```python
X_train[0]
```

```python
X_train.shape
```

```python
# running a neural network on the training and testing data, changing parameters to try to improve result

cnn_model = Sequential()


cnn_model.add(Conv2D(filters=8,            
                     kernel_size=(3, 3),       
                     activation='relu',     
                     input_shape=(28,28,1))) 

cnn_model.add(MaxPooling2D(pool_size=(2, 2)))

cnn_model.add(Conv2D(16,
                     kernel_size = (3, 3),
                     activation='relu'))

cnn_model.add(MaxPooling2D(pool_size=(2, 2)))

cnn_model.add(Flatten())

cnn_model.add(Dense(128, activation='relu'))


cnn_model.add(Dropout(0.2))

cnn_model.add(Dense(64, activation='relu'))

cnn_model.add(Dropout(0.2))

cnn_model.add(Dense(32, activation='relu'))

cnn_model.add(Dropout(0.2))

cnn_model.add(Dense(10, activation='softmax'))

cnn_model.compile(loss='categorical_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])

history = cnn_model.fit(X_train,
                        y_train,
                        batch_size=256,
                        validation_data=(X_test, y_test),
                        epochs=50,
                        verbose=1)
```

```python
#adding in the kaggel test data
test = pd.read_csv('./test.csv')
```

```python
#checking the test data
test.head()
```

```python
#reshaping the test data to fit the model
test = test.values.reshape(-1, 28, 28, 1)
```

```python
#creating the prediction from the data
preds = cnn_model.predict_classes(test)
```

```python
preds
```

```python
# turning the predictions into a kaggle submission
test_pred_df = pd.DataFrame({    
    'ImageId': range(1,test.shape[0] + 1),
    'Label': preds
})
test_pred_df.to_csv('preds.csv', index=False)
```
